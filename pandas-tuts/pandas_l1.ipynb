```python
# Create Data

# THe initial set of baby names and virth rates
names = ['Bob', 'Jessica', 'Mary', 'John', 'Mel']
births = [968, 155, 77, 578, 973]
```

```python
# We can merge them together wih the zip function
BabyDataSet = zip(names,births)
BabyDataSet
```

```python
# Now we make BabyDataSet into a DatFrame, which is like an Excel
# Spreadsheet

df = pd.DataFrame(data = BabyDataSet, columns=['Names', 'Births'])
df
```

```python
# Analyze Data

# We can do one of the following to find the most popular name:

# 1) Sort the dataframe and select the top row
# 2) Use the max() attribute to find the maximum value

# Method 1:
Sorted = df.sort(['Births'], ascending=False)
Sorted.head(1)
```

```python
# Method 2:
df['Births'].max()
```

```python
# Present Data

# Create graph
df['Births'].plot()

# Maximum value in the data set
MaxValue = df['Births'].max()

# Name associated with the maximum value
MaxName = df['Names'][df['Births'] == df['Births'].max()].values

# Text to display on graph
Text = str(MaxValue) + " - " + MaxName

# Add text to graph
plt.annotate(Text, xy=(1, MaxValue), xytext=(8, 0),
                xycoords=('axes fraction', 'data'),
                textcoords='offset points')

print "The most popular name"
df[df['Births'] == df['Births'].max()]
```
